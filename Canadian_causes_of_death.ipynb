```python
import pandas as pd 
import plotly.express as px

df = pd.read_csv('.\\data\\deaths_canada.csv')
df.info()
```

```python
df.head()
```

```python
dfn = df[(df["REF_DATE"]==df["REF_DATE"].max()) & (df['Characteristics']=="Number of deaths") & (~df['Cause of death (ICD-10)'].str.contains("Total, all causes of death")) & (df['VALUE']>0)].copy()
dfn["Age at time of death"] = dfn["Age at time of death"].str.replace("Age at time of death, ","")
dfn.rename(columns={"VALUE":"Deaths","Cause of death (ICD-10)":"Cause of death"}, inplace=True)

px.bar(dfn, x="Age at time of death",y="Deaths", hover_name="Cause of death", color="Sex",height=600)
```

```python
px.sunburst(dfn, path=["Cause of death","Age at time of death"], values="Deaths", height=600)
```

```python
px.treemap(dfn, path=["Cause of death","Age at time of death"], values="Deaths", height=800, width=800)
```
